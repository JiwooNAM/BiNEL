```python
import os, re
import numpy as np
```

```python
treats = np.load("Treatments.npy", allow_pickle=True)[:28]
print(treats)
```

```python
def safe_filename(name: str) -> str:
    name = name.strip()
    name = name.replace("/", "_")              # '/' 제거
    name = name.replace("\\", "_")             # '\' 제거
    name = re.sub(r"\s+", " ", name)           # 공백 정리
    name = name.replace(" ", "_")              # 공백 -> _
    name = re.sub(r'[<>:"|?*]', "_", name)     # 윈도우 금지문자도 제거
    return name

out_dir = "treat_pairs"   # 원하는 폴더 (없으면 자동 생성)
os.makedirs(out_dir, exist_ok=True)

for i, treat in enumerate(treats):
    arr = np.array(["DMSO", treat], dtype=object)   # 원래 treat 문자열 유지
    fname = safe_filename(treat) + ".npy"
    np.save(os.path.join(out_dir, fname), arr, allow_pickle=True)

print("done:", len(treats), "files in", out_dir)
```
