```python
from pathlib import Path

import pandas as pd
```

```python
ROOT = Path("logs/optuna/multiruns/2026-02-23_07-50-34")  # <- DATE 폴더로
EXPERIMENT = "multimodal-singleprofile-classification-CCA_random-final_test"
VERSION = "version_1"
title = "CCA_random"
```

```python
TEST_COLS = ["test/acc", "test/f1", "test/loss"]

rows = []
for run_id in range(5):
    metrics_csv = ROOT / str(run_id) / EXPERIMENT / VERSION / "metrics.csv"
    df = pd.read_csv(metrics_csv)

    last = df.iloc[-1]  # test는 마지막에만 찍힌다고 했으니
    rows.append([run_id] + [float(last[c]) for c in TEST_COLS])

table = pd.DataFrame(rows, columns=["run"] + TEST_COLS).set_index("run")
print(type(table))
print(table.to_string())

# (옵션) 저장
out_csv = ROOT / f"{title}.csv"
table.to_csv(out_csv)
print("Saved:", out_csv)
```
